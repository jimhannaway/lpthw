## LOOPS

````python
# Two types of loop in Python the For loop and the While loop.

# THE FOR LOOP
primes = [2, 3, 5, 6, 7, 9]
for unit in primes:
    print(unit)
    
# For loops can iterate over a sequence of numbers using the "range" function. 
# Note that the range function is zero based.

# Prints out the numbers 0,1,2,3,4
for x in range(5):
    print(x)

# Prints out 3,4,5
for x in range(3, 6):
    print(x)

# Prints out 3,5,7
for x in range(3, 8, 2):
    print(x)

# THE WHILE LOOP
# While loops repeat as long as a certain boolean condition is met.

# Prints out 0,1,2,3,4
count = 0
while count < 25:
    print(count)
    count += 5  # This is the same as count = count + 5

# BREAK AND CONTINUE STATEMENTS
# break is used to exit a for loop or a while loop, 
# whereas continue is used to skip the current block, and return to the "for" or "while" statement. 

# Prints out 0,1,2,3,4
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break

# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)


# THE ELSE CLAUSE IN LOOPS
# When the loop condition of "for" or "while" statement fails then code part in "else" is executed. 
# If break statement is executed inside for loop then the "else" part is skipped. 
# Note that "else" part is executed even if there is a continue statement.

# Prints out 0,1,2,3,4 and then it prints "count value reached 5"
count=0
while(count<5):
    print(count)
    count +=1
else:
    print("count value reached %d" %(count))

# Prints out 1,2,3,4,5,6
for i in range(1, 10):
    if(i%7==0): #modulo calculation if 7/7 has no remainder then break
        break
    print(i)
else:
    print("this is not printed because for loop is terminated because of break but not due to fail in condition")
    

# EXERCISE
# Loop through and print out all even numbers from the numbers list in the same order they are received. 
# Don't print any numbers that come after 237 in the sequence.

numbers = [
    951, 402, 984, 651, 360, 69, 408, 319, 601, 485, 980, 507, 725, 547, 544,
    615, 83, 165, 141, 501, 263, 617, 865, 575, 219, 390, 984, 592, 230, 105, 942, 941,
    386, 462, 47, 418, 907, 344, 236, 375, 823, 566, 597, 978, 328, 615, 953, 345,
    399, 162, 758, 219, 918, 237, 412, 566, 826, 248, 866, 950, 626, 949, 687, 217,
    815, 67, 104, 58, 512, 24, 892, 894, 767, 553, 81, 379, 843, 831, 445, 742, 717,
    958, 609, 842, 451, 688, 753, 854, 685, 93, 857, 440, 380, 126, 721, 328, 753, 470,
    743, 527
    ]
for evens in numbers:
    if evens == 237:
        break

    if evens % 2 == 1:
        continue

    print(evens)
````

## Character Classes
Character classes provide a way to match only one of a specific set of characters. A character class is created by putting the characters it matches inside **square brackets**.<br><br>
Character classes can also match ranges of characters.
Some examples:
The class [a-z] matches any lowercase alphabetic character.
The class [G-P] matches any uppercase character from G to P.
The class [0-9] matches any digit.
Multiple ranges can be included in one class. For example, [A-Za-z] matches a letter of any case.<br><br>
Place a ^ at the start of a character class to invert it.
This causes it to match any character other than the ones included.
Other metacharacters such as $ and ., have no meaning within character classes.
The metacharacter ^ has no meaning unless it is the first character in a class.<br><br>


````python
# Character class matching occurrence of characters against the set of vowels.
import re

pattern = r"[aeiou]"

if re.search(pattern, "grey"):
   print("Character Class Match 1")

if re.search(pattern, "qwertyuiop"):
   print("Character Class Match 2")

if re.search(pattern, "rhythm myths"):
   print("Character Class Match 3")

# Character class range matching for Capital Letter, Capital Letter, Any digit
# All three ranges must have a match consecutivley to obtain a match
import re

pattern = r"[A-Z][A-Z][0-9]"

if re.search(pattern, "LS8"):
   print("Class Range Match 1")

if re.search(pattern, "E3"):
   print("Class Range Match 2")

if re.search(pattern, "1abLS8"):
   print("Class Range Match 3")

# Character Class Inverted MAtch
import re

pattern = r"[^A-Z]" # find a match if any character is not a capital

if re.search(pattern, "this is all quiet"):
   print("Inverted Match 1")

if re.search(pattern, "AbCdEfG123"):
   print("Inverted Match 2")

if re.search(pattern, "THISISALLSHOUTING"):
   print("Inverted Match 3")
````

## More Meta Characters
Some more metacharacters are ***, +, ?, { and }**.  These specify numbers of repetitions.  The metacharacter __*__ means "zero or more repetitions of the previous thing". It tries to match as many repetitions as possible. The "previous thing" can be a single character, a class, or a group of characters in parentheses.<br><br>
The metacharacter **+** is very similar to *, except it means "one or more repetitions", as opposed to "zero or more repetitions".<br><br>
The metacharacter __?__ means "zero or one repetitions".<br><br>
Curly braces can be used to represent the number of repetitions between two numbers.The regex __{x,y}__ means "between x and y repetitions of something".
Hence __{0,1} is the same thing as ?__. If the first number is missing, it is taken to be zero. If the second number is missing, it is taken to be infinity.<br>
Example: 4{5,6} means "4 multiply by 5 or 6", then \1 means "repeat group 1", therefore we should match either 5*2 or 6*2 characters of "4",
i.e. 4444444444 or 444444444444 (10 or 12 "4" in the row).<br>
4{5}\0 is 44444 then 4{5}\1 is 44444 means a total = 10 fours<br>
4{6}\0 is 444444 then 4{6}\1 is 444444 means a total = 12 fours


````python
# The zero to many repetitions meta-character
import re

pattern = r"egg(spam)*"

if re.match(pattern, "egg"):
   print("Zero or One 'spam' Repetitions Match 1")

if re.match(pattern, "eggspamspamegg"):
   print("Zero or One 'spam' Repetitions Match 2")

if re.match(pattern, "spam"):
   print("Zero or One 'spam' Repetitions Match 3")

# The minimum of one up to many repetitions meta-character
import re

pattern = r"g+"

if re.match(pattern, "g"):
   print("At least one Match 1")

if re.match(pattern, "gggggggggggggg"):
   print("At least one Match 2")

if re.match(pattern, "abc"):
   print("At least one Match 3")

# The zero or one repetitions meta-character
import re

pattern = r"ice(-)?cream"

if re.match(pattern, "ice-cream"):
   print("Zero or one hyphen Match 1")

if re.match(pattern, "icecream"):
   print("Zero or one hyphen Match 2")

if re.match(pattern, "sausages"):
   print("Zero or one hyphen Match 3")

if re.match(pattern, "ice--ice"):
   print("Zero or one hyphen Match 4")

# The zero or one repetitions meta-character
import re

pattern = r"9{1,3}$"  # "$" used for search only in string using no more than 1-3 chars at a time

if re.match(pattern, "9"):
   print("Between 1 and 3 Repetitions Match 1")

if re.match(pattern, "999"):
   print("Between 1 and 3 Repetitions Match 2")

if re.match(pattern, "9999"):
   print("Between 1 and 3 Repetitions Match 3")
````

##Groups
A group can be created by surrounding part of a regular expression with parentheses. This means that a group can be given as an argument to metacharacters such as __* and ?__.<br><br>
The content of groups in a match can be accessed using the group function.
A call of __group(0) or group() returns the whole match__. A call of __group(n), where n is greater than 0, returns the nth group from the left__. The method __groups() returns all groups up from 1__.<br><br>


````python
# Groups
import re

pattern = r"egg(spam)*" # match for egg plus zro or more occurenes of the (spam)group

if re.match(pattern, "egg"):
   print("Zero or more groups Match 1")

if re.match(pattern, "eggspamspamspamegg"):
   print("Zero or more groups Match 2")

if re.match(pattern, "spam"):
   print("Zero or more groups Match 3")

# Using the group function
import re

pattern = r"a(bc)(de)(f(g)h)i"

match = re.match(pattern, "abcdefghijklmnop")
if match:
   print("Whole Group match",match.group())   # returns the whole match
   print("Whole Group match",match.group(0))  # returns the whole match
   print("Second Group match",match.group(1)) # returns 1st group of 4, count starts at first set of ()
   print("Third Group match",match.group(2))  # returns 2nd group of 4, count starts at  first set of ()
   print("Second Group Onwards match",match.groups()) #
````

## Special Sequences
There are various special sequences you can use in regular expressions. They are written as a backslash followed by another character.
One useful special sequence is a backslash and a number between 1 and 99, e.g., \1 or \17. This matches the expression of the group of that number.<br><br>
More useful special sequences are __\d, \s, and \w__.
These match digits, whitespace, and word characters respectively.
In ASCII mode they are equivalent to [0-9], [ \t\n\r\f\v], and [a-zA-Z0-9_].
In Unicode mode they match certain other characters, as well. For instance, \w matches letters with accents. Versions of these special sequences with __upper case letters - \D, \S, and \W - mean the opposite to the lower-case versions__. For instance, \D matches anything that isn't a digit.<br><br>
Additional special sequences are __\A, \Z, and \b__.
The sequences \A and \Z match the beginning and end of a string, respectively.
The sequence \b matches the empty string between \w and \W characters, or \w characters and the beginning or end of the string. <br>
\A is an anchor that says "the string starts here" so in other words no characters may occur before \A

\Z is an anchor that says "the string ends here" so in other words no characters may occur after \Z

\b matches 1 or more spaces and 0 or more letters (words). So in other words there must be at least 1 word to the left of \b and at least 1 word to the right of \b.<br>
Informally, it represents the boundary between words. The sequence \B matches the empty string anywhere else.




````python
#This simple example is helpful on special sequences.
import re

pattern = r"(.+) (.+) (.+) \1"  # match aganst any three groups of characters and the last group against the first.

# pattern = r"(.+) (.+) (.+) \2"  # match aganst any three groups of characters and the last group against the second.
# pattern = r"(.+) (.+) (.+) \3" # match aganst any three groups of characters and the last group against the third.

# if pattern = r"(.+) (.+) (.+) \1"  will print Match1
# changing the pattern to: pattern = r"(.+) (.+) (.+) \2" will print Match 2
# and then changing pattern to: pattern = r"(.+) (.+) (.+) \3"  will print Match 3
match = re.match(pattern, "abc bca cab abc")
if match:
    print ("Sequence Match 1")

match = re.match(pattern, "abc bca cab bca")
if match:
    print ("Sequence Match 2")  

match = re.match(pattern, "abc bca cab cab")
if match:
    print ("Sequence Match 3")


# Other special sequence matches
import re

pattern = r"(\D+\d)" # match one or more non-digits with any digits

match = re.match(pattern, "Hi 999!")

if match:
   print("Other 'not' Sequence Match 1")

match = re.match(pattern, "1, 23, 456!")
if match:
   print("Other 'not' Sequence Match 2")

match = re.match(pattern, " ! $?")
if match:
    print("Other 'not' Sequence Match 3")

# Other special sequence match on spaces within strings
import re

pattern = r"\b(cat)\b"  # match the word cat with a space before an after it

match = re.search(pattern, "The cat sat!")
if match:
   print ("Cat space Match 1")

match = re.search(pattern, "We s>cat<tered?")
if match:
   print ("Cat space Match 2")

match = re.search(pattern, "We scattered.")
if match:
   print ("Cat space Match 3")
````

## Regex eMail Extraction

Suppose we have a text that contains an email address:<br>
str = "Please contact info@sololearn.com for assistance"<br>

Our goal is to extract the substring "info@sololearn.com".<br><br>
A basic email address consists of a word and may include dots or dashes. This is followed by the @ sign and the domain name (the name, a dot, and the domain name suffix).<br><br>
This is the basis for building our regular expression.<br>
pattern = r"([\w\\.-]+)@([\w\\.-]+)(\.[\w\\.]+)"

[\w\.-]+ matches one or more word character, dot or dash.<br><br>
The regex above says that the string should contain a word (with dots and dashes allowed), followed by the @ sign, then another similar word, then a dot and another word.


````python
import re

pattern = r"([\w\.-]+)@([\w\.-]+)(\.[\w\.]+)"

str = "Please contact jim_hannaway@pythoncoders.co.uk for assistance"

match = re.search(pattern, str)
if match:
   print(match.group())
````
